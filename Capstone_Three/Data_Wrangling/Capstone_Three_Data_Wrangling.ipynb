#### Creation of Training and Test Data
The dataset considered for Capstone Three proejct contains huge number of images and it will be computationally difficult to train and fit the model on such a huge dataset.
However the model creation and training will be performed in Colab notebook, it still contains huge dataset.
Hence this notebook will create the training and test dataset folders.
The absic approach used to create training and test dataset is as follows:
1. If a particular folder contains more than 200 images then first 100 images will be used as training data and next 100 wil be used as test data.
2. If the folder contains less than 200 images but greater than 50 images then it will be split in the ratio of 80% fro training and 20% for test.
3. If the folder consists of less than 50 images then it will not considered for classification.

```python
# Importing os library to read and write files and folders
import os
import shutil
```

```python
os.getcwd()
```

```python
main_folder = 'D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\extracted_images'
# Make sure it is the same directory
os.listdir(main_folder)
```

```python
for subfolder in os.listdir(main_folder):
    num_images = len(os.listdir(main_folder+'\\'+subfolder))
    if num_images >= 200:
        os.chdir('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\train')
        os.makedirs(subfolder, exist_ok=True)
        os.chdir('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\test')
        os.makedirs(subfolder, exist_ok=True)
        i_train=100
        i_test=200
        for i in range(0,i_train):
            shutil.copy(main_folder+'\\'+subfolder+'\\'+os.listdir(main_folder+'\\'+subfolder)[i],
                        ('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\train\\'+subfolder))
        for i in range(i_train, i_test):
            shutil.copy(main_folder+'\\'+subfolder+'\\'+os.listdir(main_folder+'\\'+subfolder)[i], 
                        ('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\test\\'+subfolder))
    if (num_images < 200) and (num_images>=50):
        os.chdir('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\train')
        os.makedirs(subfolder)
        os.chdir('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\test')
        os.makedirs(subfolder)
        i_train=int(0.8*num_images)
        i_test=num_images
        for i in range(0,i_train):
            shutil.copy(main_folder+'\\'+subfolder+'\\'+os.listdir(main_folder+'\\'+subfolder)[i], 
                        ('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\train\\'+subfolder))
        for i in range(i_train, i_test):
            shutil.copy(main_folder+'\\'+subfolder+'\\'+os.listdir(main_folder+'\\'+subfolder)[i], 
                        ('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\test\\'+subfolder))
```

```python
for subfolder in os.listdir('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\train'):
    print('Train:', subfolder, ':', len(os.listdir('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\train\\'+subfolder)))
    print('Test:', subfolder, ':', len(os.listdir('D:\\Springboard\\Technical Project\\23_Capstone Three\\data\\test\\'+subfolder)))
```
